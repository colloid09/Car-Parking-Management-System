```python
#write a python program to find the factorial of the number using recursion

def fact(n):
  if n==0 or n==1:
    return 1
  else:
    return n*fact(n-1)

n = int(input("Enter a number: "))
fact(n)
```

```python
#write a python program to insert an element into the sorted list
def sortlist(l,ele):
  for i in range(len(l)):
    if (ele<=l[i]):
      l.insert(i,ele)
      break
  else:
    l.append(ele)
  print(l)

l=[2,1,4,3,5]
l.sort()
print("Sorted Elements ",l)
ele = int(input("Enter the element to insert: "))
sortlist(l,ele)
```

```python
from os import name
#write a python program using object oriented programming to demonstrate encaplusation, inheritance and method overriding

class Person:
  def __init__(self,name,add,pno):
    self.__name=name
    self.__add=add
    self.__pno=pno
  def display(self):
    print("---Employee Details---")
    print("Name: ",self.__name)
    print("Address: ",self.__add)
    print("Contact: ",self.__pno)

class Employee(Person):
  def __init__(self,name,add,pno,eno,sal):
    super().__init__(name,add,pno)
    self._eno=eno
    self._sal=sal
  def display(self):
    super().display()
    print("Employee Number: ",self._eno)
    print("Salary: ",self._sal)

class Manager(Employee):
  def __init__(self,name,add,pno,eno,sal):
    super().__init__(name,add,pno,eno,sal)
    self.calculate()

  def calculate(self):
    self.__da=self._sal*0.2
    self.__hra=self._sal*0.1
    self.__gross=self._sal+self.__da+self.__hra

  def display(self):
    super().display()
    print("DA: ",self.__da)
    print("HRA: ",self.__hra)
    print("Gross Salary: ",self.__gross)

name = input("Enter name: ")
add = input("Enter address: ")
pno = int(input("Enter contact number: "))
eno = int(input("Enter employee number: "))
sal = int(input("Enter salary: "))
manager1 = Manager(name,add,pno,eno,sal)
manager1.display()
```

```python
#write a python program to check whether a given string is palindrome or not
def palindrome(s):
  if s==s[::-1]:
    print("It is Palindrome")
  else:
    print("Not Palindrome")

s=input("Enter a string: ")
palindrome(s)
```

```python
#write a python program to check whether a given string is palindrome or not

def palindrome(s):
  for i in range(len(s)):
    if s[i]!=s[len(s)-i-1]:
      print("Not a Palindrome")
      break
  else:
    print("It is a Palindrome")

s=input("Enter a string: ")
palindrome(s)
```

```python
#plotting

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.2)
print("X=", x)

y_sine = np.sin(x)
y_cosine = np.cos(x)

plt.plot(x, y_sine)
plt.plot(x, y_cosine)
plt.xlabel('x values')
plt.ylabel('y sine and cosine values')
plt.title('Sine and Cosine Functions')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
#Implement a python program to demonstrate
#1) Importing Datasets
#2) Cleaning the Data
#3) Data frame manipulation using Pandas.

#3) Data frame manipulation using Pandas.
import pandas as pd
d1 ={'city':['mangalore','bangalore','mysore'],'temp':[35,30,29]}
d2 ={'city':['mangalore','bangalore','mysore'],'humidity':[60,50,40]}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
print(df1)
print(df2)

#Add row
print("\nAdd row")
df1.loc[5]=['Belgavi',37]
print(df1)

#drop row
print("\nDrop row")
df1=df1.drop(5)
print(df1)

#merging
print("\nMerge")
pmerge=pd.merge(df1,df2,on='city')
print(pmerge)

#concat
print("\nConcat")
df3=pd.concat([df1,df2])
print(df3)

#concat with axis
print("\nConcat with axis")
df3=pd.concat([df1,df2],axis=1)
print(df3)

#sorting dataframes
print("\nSort")
df1.sort_values(by='temp', inplace=True)
print(df1)
```
